```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from setups import get_local_safe_setup
import os
import time

driver = get_local_safe_setup()

# Call url through selenium driver
driver.get("https://www.daft.ie/property-for-sale/ireland")
#driver.get("https://www.daft.ie/property-for-sale/dublin-city")

#accept_button = driver.find_element_by_xpath("//button[contains(text(), 'Accept All Cookies')]")
accept_button = driver.find_element(By.XPATH, '//*[@id="js-cookie-modal-level-one"]/div/main/div/button[2]')
if accept_button.is_displayed():
    accept_button.click()
```

```python
props = []

for page in range(10):

    results = driver.find_elements(By.CSS_SELECTOR, ".djuMQD")

    for result in results:
        link = result.find_element(By.TAG_NAME, "a").get_attribute("href")
        
        if link not in props:
            props.append(link)

    print(f"{len(results)} Results in Page {page+1}")
    
    next_but = driver.find_element(By.XPATH, '//*[@aria-label="Next >"]')
    next_but.click()
    time.sleep(1)
    
print(f"{len(props)} Properties")
```

```python
addresses =[]
countries = []
agents = []
prop_types = []
units = []
prices = []

p=1
for prop in props:
    driver.get(prop)
    print(f"Scrape Property {p}")
    p+=1
    countries.append(driver.find_element(By.TAG_NAME, "h1").text.split(",")[-1])
    addresses.append(driver.find_element(By.TAG_NAME, "h1").text)
    agents.append(driver.find_element(By.CSS_SELECTOR, '.boRqb').text)
    
    try:
        #One Property    
        prop_types.append(driver.find_element(By.XPATH, '//*[@data-testid="property-type"]').text)
        units.append(1)
        prices.append(driver.find_element(By.TAG_NAME, "h2").text) 
        
    except:
        #New Developments     
        prop_types.append(driver.find_element(By.CSS_SELECTOR, '.eKVlRq').text)
        units.append(driver.find_element(By.CSS_SELECTOR, '.ikMOXo').text.split()[0])
        #units.append(driver.find_element(By.CSS_SELECTOR, ".styles__InfoSection-sc-15fxapi-7").text.split()[0])                                            
        prices.append(driver.find_element(By.CSS_SELECTOR, ".icGLog").text) 
    
    time.sleep(1)
    
df = pd.DataFrame({
    
    "Country":countries, "Agent":agents, "Address":addresses, "PropertyType":prop_types, "Price":prices, "Units":units
})

df.to_csv("Properities.csv", index=False)
df
```
